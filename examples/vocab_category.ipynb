```python
from AnkiTools.tools.read import readApkg
```

```python
anki = readApkg("/Users/patarapolw/Downloads/cn_category.apkg")
```

```python
tag_dict = dict()
for note in anki.notes.values():
    for tag in note['tags']:
        tag_dict[tag] = tag_dict.setdefault(tag, 0) + 1
tag_dict
```

```python
import operator
sorted(tag_dict.items(), key=operator.itemgetter(1))
```

```python
tag_list = []
for k, v in tag_dict.items():
    if 10 < v < 100 and 'manual' not in k:
        tag_list.append(k)
tag_list
```

```python
import re
from bs4 import BeautifulSoup

for tag in tag_list:
    print(tag)
    for note in anki.notes.values():
        if tag in note['tags']:
            content = note["content"]
            print(content[0] + '\t' + re.sub('\[.+\]', '', content[2]) + '\t' 
                  + BeautifulSoup(content[1], 'html.parser').text)
    print()
```
